```python
import pandas as pd

# Load the mutual fund master dataset
# Adjust the filename if your actual raw CSV is named differently
file_path = '../data/raw/01_fund_master.csv'

try:
    df_master = pd.read_csv(file_path)
    
    print("=" * 50)
    print("📊 FUND MASTER EDA: UNIQUE VALUES")
    print("=" * 50)
    
    # 1. Unique Fund Houses (AMCs)
    fund_houses = df_master['fund_house'].dropna().unique()
    print(f"\n🏢 Unique Fund Houses ({len(fund_houses)} total):")
    # Printing just the first 10 so it doesn't flood your screen
    print(fund_houses[:10], "... [truncated]") 
    
    # 2. Unique Categories
    categories = df_master['category'].dropna().unique()
    print(f"\n📁 Unique Categories ({len(categories)} total):")
    print(categories)
    
    # 3. Unique Sub-categories
    sub_cats = df_master['sub_category'].dropna().unique()
    print(f"\n📂 Unique Sub-categories ({len(sub_cats)} total):")
    print(sub_cats[:10], "... [truncated]")
    
    # 4. Unique Risk Grades
    risks = df_master['risk_category'].dropna().unique()
    print(f"\n⚠️ Unique Risk Grades ({len(risks)} total):")
    print(risks)

except FileNotFoundError:
    print(f"Error: Could not find {file_path}. Please ensure the file exists and the name matches.")
```

```python
import pandas as pd

# Define our file paths based on your folder structure
master_path = '../data/raw/01_fund_master.csv'
nav_path = '../data/raw/02_nav_history.csv'

try:
    # Load the datasets
    df_master = pd.read_csv(master_path)
    df_nav = pd.read_csv(nav_path)
    
    print("=" * 60)
    print("✅ TASK 7: DATA QUALITY REPORT - AMFI CODE VALIDATION")
    print("=" * 60)
    
    # We know from earlier that the master file uses 'amfi_code'
    master_id_col = 'amfi_code'
    
    # Check if NAV history uses 'scheme_code' or 'amfi_code'
    nav_id_col = 'scheme_code' if 'scheme_code' in df_nav.columns else 'amfi_code'
    
    # Extract unique codes as mathematical sets for easy comparison
    master_codes = set(df_master[master_id_col].dropna().unique())
    nav_codes = set(df_nav[nav_id_col].dropna().unique())
    
    # Calculate overlaps and differences
    overlap = master_codes.intersection(nav_codes)
    missing_in_nav = master_codes - nav_codes
    missing_in_master = nav_codes - master_codes
    
    # Print the Data Quality Report
    print(f"Total Unique AMFI Codes in Fund Master:  {len(master_codes)}")
    print(f"Total Unique Codes in NAV History:       {len(nav_codes)}")
    print("-" * 60)
    print(f"✔️ Codes present in BOTH datasets:       {len(overlap)}")
    print(f"⚠️ Codes in Master but MISSING in NAV:   {len(missing_in_nav)}")
    print(f"ℹ️ Codes in NAV but MISSING in Master:   {len(missing_in_master)}")
    print("=" * 60)
    
    # Final Summary
    if len(missing_in_nav) == 0:
        print("\nPerfect Match! All master funds have historical NAV data. Data quality is excellent.")
    else:
        print(f"\nData Quality Alert: You have {len(missing_in_nav)} funds in your master list that lack pricing history.")
        print("Example missing codes:", list(missing_in_nav)[:5])

except Exception as e:
    print(f"An error occurred: {e}")
```
